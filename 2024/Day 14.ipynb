In this example, the quadrants contain 1, 3, 4, and 1 robot. Multiplying these together gives a total safety factor of 12.

Predict the motion of the robots in your list within a space which is 101 tiles wide and 103 tiles tall. What will the safety factor be after exactly 100 seconds have elapsed?



```python
import numpy as np
import re
```

```python
WIDTH = 101
HEIGHT = 103

class Robot:
    def __init__(self, pos: np.array, vel: np.array, ledger):
        self.pos = pos
        self.vel = vel
        self.ledger = ledger

    def __repr__(self):
        return f'Robot {self.pos=} {self.vel=}'

    def move(self):
        self.pos = self.pos + self.vel

        # code reliant on all velocities being within WIDTH/HEIGHT (so no multiple wraps in one move):
        if self.pos[0] < 0: 
            self.pos[0] = WIDTH + self.pos[0]
        elif self.pos[0] >= WIDTH:
            self.pos[0] -= WIDTH
        if self.pos[1] < 0:
            self.pos[1] = HEIGHT + self.pos[1]
        elif self.pos[1] >= HEIGHT:
            self.pos[1] -= HEIGHT

    @property
    def quadrant(self): # valid for HEIGHT/WIDTH uneven 
        if self.pos[0] < WIDTH // 2: # leftie
            if self.pos[1] < HEIGHT // 2: # upper
                return 0
            elif self.pos[1] > HEIGHT // 2 : # down
                return 1 
        elif self.pos[0] > WIDTH // 2: # rightie
            if self.pos[1] < HEIGHT // 2: # upper
                return 2
            elif self.pos[1] > HEIGHT // 2: # down
                return 3
        return None # middle


    @staticmethod
    def grid_move(ledger):
        for robot in ledger:
            robot.move()

    @staticmethod
    def quadrant_count(ledger):
        counts = [0,0,0,0]
        for robot in ledger:
            if robot.quadrant is not None:
                counts[robot.quadrant] += 1
        return counts
```

```python
with open('input14.txt') as file:
    ledger = set()
    for line in file:
        pos_x, pos_y, vel_x, vel_y = re.findall(r"-?\d+", line)
        pos = np.array((pos_x, pos_y), dtype=int)
        vel = np.array((vel_x, vel_y), dtype=int)
        ledger.add(Robot(pos, vel, ledger))
next(iter(ledger))
```

```python
for _ in range(100):
    Robot.grid_move(ledger)
counts = Robot.quadrant_count(ledger)
safety_factor = np.prod(counts)
safety_factor
```
