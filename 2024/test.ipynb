```python
with open('input5.txt') as file:
    value_before_key = {}
    for line in file:
        if "|" not in line:
            break
        value, key = (int(x) for x in line.split('|')) 
        value_before_key.setdefault(key, set()).add(value)
    file.seek(0)
    updates = [tuple(int(x) for x in line.split(",")) for line in file if "," in line]
value_before_key
```
