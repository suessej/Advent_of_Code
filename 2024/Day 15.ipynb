The lanternfish use their own custom Goods Positioning System (GPS for short) to track the locations of the boxes. The GPS coordinate of a box is equal to 100 times its distance from the top edge of the map plus its distance from the left edge of the map. (This process does not stop at wall tiles; measure all the way to the edges of the map.)

So, the box shown below has a distance of 1 from the top edge of the map and 4 from the left edge of the map, resulting in a GPS coordinate of 100 * 1 + 4 = 104.

The lanternfish would like to know the sum of all boxes' GPS coordinates after the robot finishes moving. In the larger example, the sum of all boxes' GPS coordinates is 10092. In the smaller example, the sum is 2028.

Predict the motion of the robot and boxes in the warehouse. After the robot is finished moving, what is the sum of all boxes' GPS coordinates?

```python
"""
Data structure:
An array (list of lists) with one list per line in the grid and number entries for the fields occupied by a box
A similar array for walls
For each move we iteratively look forward to see if the space is empty, a wall or box (iterate if box)
if move we move robot position and we move boxes (just need to remove last box and add one new box in the empty space to move into)
"""
```

```python
with open('input15.txt') as file:
    boxes = []
    walls = []
    robot_moves = []
    for x, line in enumerate(file):
        if line.startswith('#'):
            # grid:
            boxes.append([])
            walls.append([])
            for y, pos in enumerate(line):
                if pos == '#':
                    walls[-1].append(y)
                elif pos == 'O':
                    boxes[-1].append(y)
                elif pos == '@':
                    rob_pos = (x, y)

        else:
            for m in line:
                if m == '>' or m == '<' or m == 'v' or m == '^':
                    robot_moves.append(m)
grid_size = len(boxes)
len(boxes), len(walls[0]), robot_moves[:10], walls[1], rob_pos
```

```python
def move(orig, dir):
    x, y = orig
    if dir == '<':
        y -= 1
    elif dir == '>':
        y += 1
    elif dir == '^':
        x -= 1
    elif dir == 'v':
        x += 1
    return (x, y)


def look_ahead(orig, dir):
    """returns pos of next empty space, None if no space (wall)"""
    x, y = move(orig, dir)
    if y in walls[x]:
        return None
    elif y in boxes[x]:
        return look_ahead((x,y), dir)
    else: # space
        return (x, y)
    
look_ahead((26,24), 'v')
```

```python
for m in robot_moves:
    next_space = look_ahead(rob_pos, m)
    if next_space:
        rob_pos = move(rob_pos, m)
        if rob_pos != next_space:
            boxes[rob_pos[0]].remove(rob_pos[1])
            boxes[next_space[0]].append(next_space[1])
```

```python
def gps(pos):
    x, y = pos
    return 100 * x + y

gps_sum = 0
for x, line in enumerate(boxes):
    for y in line:
        gps_sum += gps((x, y))
gps_sum
```

```python
with open('input15.txt') as file:
    boxes = []
    walls = []
    for x, line in enumerate(file):
        if line.startswith('#'):
            # grid:
            boxes.append(set())
            walls.append(set())
            for y, pos in enumerate(line):
                y *= 2
                if pos == '#':
                    walls[-1].update([y, y+1])
                elif pos == 'O':
                    boxes[-1].add(y) # the new double spaced boxes are stored with their starting index
                elif pos == '@':
                    rob_pos = (x, y)
```

```python
"""
Our old efficient approach will not work here. We need to find the whole cascade of boxes affected. If a single one cannot move nothing moves. If we can move, everything moves at once.
"""
```

```python
for dir in robot_moves:
    queue = [move(rob_pos, dir)]
    cascade = set()
    movable = True
    while queue:
        x, y = queue.pop()
        if y in walls[x]:
            movable = False
            break
        
        # Box check:
        b_y = None
        if y-1 in boxes[x]:
            b_y = y-1
        elif y in boxes[x]:
            b_y = y
        if b_y is not None and (x, b_y) not in cascade:
            new_box = (x, b_y)
            cascade.add(new_box)
            if dir == '>':
                queue.append((x, b_y + 2))
            else:
                queue.append(move(new_box, dir))
            if dir == '^' or dir == 'v':
                queue.append(move((x, b_y + 1), dir))

    if movable:
        rob_pos = move(rob_pos, dir)
        for x, y in cascade:
            boxes[x].remove(y)
        for x, y in cascade:
            if dir == '^':
                boxes[x - 1].add(y)
            elif dir == 'v':
                boxes[x+1].add(y)
            elif dir == '<':
                boxes[x].add(y-1)
            elif dir == '>':
                boxes[x].add(y+1)
```

```python
gps_sum = 0
for x, line in enumerate(boxes):
    for y in line:
        gps_sum += gps((x, y))
gps_sum
```
