```python
with open("input18.txt") as file:
    data = [line.split(",") for line in file]
    data = [(int(x), int(y)) for x, y in data]
```

```python
kilobyte = set(data[:1024])
len(data)
```

```python
from collections import deque
start = (0,0)
end = (70,70)

def neighbors(loc):
    i, j = loc
    candidates = [(i, j+1), (i, j-1), (i+1, j), (i-1,j)]

    return [(i,j) for (i,j) in candidates if 0 <= i <= end[0] and 0 <= j <= end[0]]

# FIFO queue:
dist = {}
dist[start] = 0
visited = set()
queue = deque()
queue.append(start)
while queue:
    tile = queue.popleft()
    if tile not in visited:
        visited.add(tile)
        for n in neighbors(tile):
            if n not in kilobyte and n not in dist:
                dist[n] = dist[tile] + 1
                queue.append(n)
                if n == end:
                    queue = deque()
                    print(f"end reached in {dist[n]} steps")
                    break
    
```

```python
def end_reachable(obstacles):
    visited = set()
    queue = []
    queue.append(start)
    while queue:
        tile = queue.pop()
        visited.add(tile)
        for n in neighbors(tile):
            if n not in visited and n not in obstacles:
                queue.append(n)
                if n == end:
                    return True
    return False
end_reachable(data[:-1000])
```

```python
lower_bound = 1024
upper_bound = len(data)

while True:
    if upper_bound - lower_bound < 2:
        break
    cand_bound = (upper_bound + lower_bound) // 2
    obstacles = set(data[:cand_bound])
    if end_reachable(obstacles):
        lower_bound = cand_bound
    else:
        upper_bound = cand_bound
lower_bound, upper_bound, data[upper_bound-1]
```
