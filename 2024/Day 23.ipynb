```python
from collections import defaultdict
pairwise = defaultdict(set)
for comp1, comp2 in sample_data:
    pairwise[comp1].add(comp2)
    pairwise[comp2].add(comp1)
```

```python
from itertools import combinations

triples = set()
visited = set()
for c1, cnc in pairwise.items():
    for c2, c3 in combinations(cnc, 2):
        if c2 not in visited and c3 not in visited:
            if c3 in pairwise[c2]:
                triples.add(frozenset([c1, c2, c3]))
    visited.add(c1)
```

Since it doesn't seem like any employees are around, you figure they must all be at the LAN party. If that's true, the LAN party will be the largest set of computers that are all connected to each other. That is, for each computer at the LAN party, that computer will have a connection to every other computer at the LAN party.

In the above example, the largest set of computers that are all connected to each other is made up of co, de, ka, and ta. Each computer in this set has a connection to every other computer in the set:

The LAN party posters say that the password to get into the LAN party is the name of every computer at the LAN party, sorted alphabetically, then joined together with commas. (The people running the LAN party are clearly a bunch of nerds.) In this example, the password would be co,de,ka,ta.

What is the password to get into the LAN party?

```python
max_pairs = max(len(coll) for coll in pairwise.values()) + 1
max_pairs
```

```python
len({k: v for k, v in pairwise.items() if len(v) == 13}) # all 520 computers have 13 connections exactly
```

```python
# since the solution requires a unique largest LAN we can stop early
def find_lan(n):
    visited = set()
    counter = 0
    for c1, cnc in pairwise.items():
        for draw in combinations(cnc, n-1):
            if all(c not in visited for c in draw):
                if all(c3 in pairwise[c2] for c2, c3 in combinations(draw, 2)):
                    return set([*draw, c1])
        visited.add(c1)
find_lan(13)
```

```python
for n in range(max_pairs, 0, -1):
    if largest_set := find_lan(n):
        break
largest_set
```

```python
solution = ','.join(sorted(largest_set))
solution
```
