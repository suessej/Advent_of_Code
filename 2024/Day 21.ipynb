The robot has no problem navigating the ship and finding the numeric keypad, but it's not designed for button pushing: it can't be told to push a specific button directly. Instead, it has a robotic arm that can be controlled remotely via a directional keypad.

The directional keypad has two rows of buttons: a gap / ^ (up) / A (activate) on the first row and < (left) / v (down) / > (right) on the second row. Visually, they are arranged like this:

    +---+---+

When the robot arrives at the numeric keypad, its robotic arm is pointed at the A button in the bottom right corner. After that, this directional keypad remote control must be used to maneuver the robotic arm: the up / down / left / right buttons cause it to move its arm one button in that direction, and the A button causes the robot to briefly move forward, pressing the button being aimed at by the robotic arm.

For example, to make the robot type 029A on the numeric keypad, one sequence of inputs on the directional keypad you could use is:

< to move the arm from A (its initial position) to 0.
A to push the 0 button.
^A to move the arm to the 2 button and push it.
>^^A to move the arm to the 9 button and push it.
vvvA to move the arm to the A button and push it.
In total, there are three shortest possible sequences of button presses on this directional keypad that would cause the robot to type 029A: <A^A>^^AvvvA, <A^A^>^AvvvA, and <A^A^^>AvvvA.

Unfortunately, the area containing this directional keypad remote control is currently experiencing high levels of radiation and nobody can go near it. A robot needs to be sent instead.

When the robot arrives at the directional keypad, its robot arm is pointed at the A button in the upper right corner. After that, a second, different directional keypad remote control is used to control this robot (in the same way as the first robot, except that this one is typing on a directional keypad instead of a numeric keypad).

There are multiple shortest possible sequences of directional keypad button presses that would cause this robot to tell the first robot to type 029A on the door. One such sequence is v<<A>>^A<A>AvA<^AA>A<vAAA>^A.

Unfortunately, the area containing this second directional keypad remote control is currently -40 degrees! Another robot will need to be sent to type on that directional keypad, too.

There are many shortest possible sequences of directional keypad button presses that would cause this robot to tell the second robot to tell the first robot to eventually type 029A on the door. One such sequence is <vA<AA>>^AvAA<^A>A<v<A>>^AvA^A<vA>^A<v<A>^A>AAvA^A<v<A>A>^AAAvA<^A>A.

The Historians are getting nervous; the ship computer doesn't remember whether the missing Historian is trapped in the area containing a giant electromagnet or molten lava. You'll need to make sure that for each of the five codes, you find the shortest sequence of button presses necessary.

The complexity of a single code (like 029A) is equal to the result of multiplying these two values:

The length of the shortest sequence of button presses you need to type on your directional keypad in order to cause the code to be typed on the numeric keypad; for 029A, this would be 68.
The numeric part of the code (ignoring leading zeroes); for 029A, this would be 29.
In the above example, complexity of the five codes can be found by calculating 68 * 29, 60 * 980, 68 * 179, 64 * 456, and 64 * 379. Adding these together produces 126384.

Find the fewest number of button presses you'll need to perform in order to cause the robot in front of the door to type each code. What is the sum of the complexities of the five codes on your list?

```python
"""
Initial Notes:
<vA<AA>>^AvAA<^A>A<v<A>>^AvA^A<vA>^A<v<A>^A>AAvA^A<v<A>A>^AAAvA<^A>A
  v <<   A >>  ^ A   <   A > A  v  A   <  ^ AA > A   < v  AAA >  ^ A
         <       A       ^   A     >        ^^   A        vvv      A
                 0           2                   9                 A


What I am thinking about is: What makes a path optimal? What are the constraints?

On the highest level (numpad) the sequence is predetermined. 
On each subsequent level, given the required input sequence for the higher level, we know exactly which inputs we need, but there can be some variability regarding the order of inputs (eg do i go v< or <v to move from 9 to 5)

This variability can compound over multiple level, such as the difference between Av<A and A<vA

However, each "A" provides a natural reset where all lower level robots will need to press "A" as well. Thus we can chunk sequences where all robots start on A and end on A. We can compute costs for simple sequences and chain those to compute more elaborate, multi-level sequences.

In the end we only need to compute one optimal sequence for each num key sequence (start on button X, end with pressing button Y) since all higher level keypads start and end on A

Im gonna assume that an ordered sequence (eg vv>>) will always be weakly better than unordered (eg v>v>)
"""
```

```python
def coord_to_seq(start, end, gap=None):
    s_x, s_y = start
    e_x, e_y = end

    hor = vert = ""

    if e_x > s_x:
        vert = "v" * abs(e_x - s_x)
    elif s_x > e_x:
        vert = "^" * abs(e_x - s_x)
    if e_y > s_y:
        hor = ">" * abs(s_y - e_y)
    elif e_y < s_y:
        hor = "<" * abs(s_y - e_y)

    hv = hor + vert
    vh = vert + hor

    if gap:
        g_x, g_y = gap
        if s_x == g_x and e_y == g_y:
            return [vh]
        elif s_y == g_y and e_x == g_x:
            return [hv]

    if hv != vh:
        return [vh, hv]
    else:
        return [vh]
coord_to_seq((0,0), (3,5), (3,0))
```

```python
def num_to_key(start, end):
    coord = {"7" : (0,0),
             "8" : (0,1),
             "9" : (0,2),
             "4" : (1,0),
             "5" : (1,1),
             "6" : (1,2),
             "1" : (2,0),
             "2" : (2,1),
             '3' : (2,2),
             "gap" : (3,0),
             '0'   : (3,1),
             "A"   : (3,2),}
    
    s = coord[start]
    e = coord[end]
    gap = coord["gap"]
    paths = coord_to_seq(s, e, gap)
    return [path + "A" for path in paths]

def key_to_key(start,end):
    coord = {"gap" : (0,0),
             "^" : (0,1),
             "A" : (0,2),
             "<" : (1,0),
             "v" : (1,1),
             ">" : (1,2),
             }
    s = coord[start]
    e = coord[end]
    gap = coord["gap"]
    paths = coord_to_seq(s, e, gap)
    return [path + "A" for path in paths]

num_to_key("5","9"), key_to_key("A", "v")
```

```python
from itertools import product

def ktk_exh(seq):
    path_list = []
    for start, end in zip("A" + seq, seq):
        path_list.append(key_to_key(start, end))
    return ["".join(p) for p in product(*path_list)]

def calc_opt(start, end):
    key1 = num_to_key(start, end)
    key2 = [ktk_exh(x) for x in key1]
    key2 = [string for sublist in key2 for string in sublist]
    key3 = [ktk_exh(x) for x in key2]
    key3 = [string for sublist in key3 for string in sublist]
    return min(key3, key=len)
```

```python
num_buttons = "0123456789A"
opt_seq = {}
for start in num_buttons:
    for end in num_buttons:
        opt_seq[start + end] = calc_opt(start, end)
opt_seq
```

```python
code = '980A'
btn_presses = ""
for pair in zip("A" + code, code):
    pair = "".join(pair)
    btn_presses += opt_seq[pair]
btn_presses, len(btn_presses)
```

```python
puzzle_input = """973A
836A
780A
985A
413A"""
codes = puzzle_input.splitlines()
codes
```

```python
sum_compl = 0
for code in codes:
    btn_presses = ""
    for pair in zip("A" + code, code):
        pair = "".join(pair)
        btn_presses += opt_seq[pair]
    num_code = int(code[:3])
    compl = num_code * len(btn_presses)
    sum_compl += compl
sum_compl
```

Just as the missing Historian is released, The Historians realize that a second member of their search party has also been missing this entire time!

A quick life-form scan reveals the Historian is also trapped in a locked area of the ship. Due to a variety of hazards, robots are once again dispatched, forming another chain of remote control keypads managing robotic-arm-wielding robots.

This time, many more robots are involved. In summary, there are the following keypads:

One directional keypad that you are using.
25 directional keypads that robots are using.
One numeric keypad (on a door) that a robot is using.
The keypads form a chain, just like before: your directional keypad controls a robot which is typing on a directional keypad which controls a robot which is typing on a directional keypad... and so on, ending with the robot which is typing on the numeric keypad.

The door codes are the same this time around; only the number of robots and directional keypads has changed.

Find the fewest number of button presses you'll need to perform in order to cause the robot in front of the door to type each code. What is the sum of the complexities of the five codes on your list?

```python
"""
We will need an efficient and robust algorithm for this part. 

Let's concentrate on the keypad to keypad conversion first. I want to create a dictionary for the optimal one level expansion for every sequence (start position) - (end position) - (press A). Eg if the arm starts at "A", needs to move to "v" and then activate the expansion sequence is "<vA" or "v<A". Since every such sequence ends on "A", the optimal expansion does not depend on outside context - the only thing we need is to specify start and end. 

For many sequences, there is really only one possible expansion, eg start(<), end(v) only ever expands into >A. For those with multiple values, we will have to figure out what's going on to ensure our dict will have the optimal one. For instance, arranging the sequence on the next or subsequent levels such that buttons farther from A are pressed in succession provides speedups. 

Once we have this dictionary, we can simply count the expansions, eg if we have one A -> v -> press passage in our full sequence, we can use the dict to simply count how many other x->y->press couplings we will have at the next level and so on.
"""
```

```python
ktk_expansions = {}
for start in "<>v^A":
    for end in "<>v^A":
        expansions = key_to_key(start, end)
        if len(expansions) == 1:
            ktk_expansions[start+end] = expansions[0]
        else:
            print(f"{start}{end}: {expansions}")
```

```python
# All of these kinda come down to: If I have to go horizontal and vertical, which direction should i go first?
# On the next level, we want to go left before going up/down, since this means we can move << from A
ktk_expansions[">^"] = '<^A'
ktk_expansions["Av"] = '<vA'

# This leaves ^>: ['v>A', '>vA'] as well as vA: ['^>A', '>^A']
# With ^>, the expansion v>A is preferable. Why? In order to visit v, > and A we have to move left, down, right and up exactly once (left + down to reach v, before going back up to A. We hit up > somewhere in between). As said earlier, if we have to go left, it is good to do it immediately (coming from A), since this allows us to double tap << on the expansion.
ktk_expansions["^>"] = 'v>A'

# (Note: I did some more exploration to find the correct order for the last three, but omitted that here to make the notebook a liiitle bit leaner)
# The last holdous is vA: ['^>A', '>^A']
```

```python
# Let's write code that finds optimal expansions for the last case (expanding vA)

def expand_seq(seq, exp_vA, init="A"):
    expansion = ""
    for pair in zip(init + seq, seq):
        pair = "".join(pair)
        if pair == 'vA':
            expansion += exp_vA
        else:
            expansion += ktk_expansions[pair]
    return expansion
```

```python
# let's clean up the expand function, we're gonna need it:
def expand_seq(seq, init="A"):
    expansion = ""
    for pair in zip(init + seq, seq):
        pair = "".join(pair)
        expansion += ktk_expansions[pair]
    return expansion
```

```python
# Now that we finally have a complete key_to_key dictionary...let's make our num to key dictionary
ntk_expansions = {}
for start in "0123456789A":
    for end in "0123456789A":
        expansions = num_to_key(start, end)
        if len(expansions) == 1:
            ntk_expansions[start+end] = expansions[0]
        else:
            undecided = True
            mutated = expansions.copy()
            while undecided:
                mutated = [expand_seq(m) for m in mutated]
                m_len = [len(seq) for seq in mutated]
                if m_len.count(min(m_len)) == 1:
                    undecided = False
                    min_idx = m_len.index(min(m_len))
            ntk_expansions[start+end] = expansions[min_idx]
print(ntk_expansions)
```

```python
# Now we are ready for the final push:
# First we prep our ktk_mappings dict from ktk_expansions for slightly easier handling later:
ktk_mappings = {}
for seed, expa in ktk_expansions.items():
    mappings = ["".join(pair) for pair in zip("A"+expa, expa)]
    ktk_mappings[seed] = mappings

ntk_mappings = {}
for seed, expa in ntk_expansions.items():
    mappings = ["".join(pair) for pair in zip("A"+expa, expa)]
    ntk_mappings[seed] = mappings

ktk_expansions['^<'], ktk_mappings['^<'], ntk_expansions["04"], ntk_mappings["04"]
```

```python
def num_press(code, num_keypads):
    seq_count = dict.fromkeys(ktk_expansions, 0)

    # get initial seq_counts going (num to keypad conversion):
    for pair in zip("A" + code, code):
        pair = "".join(pair)
        for keypad_pair in ntk_mappings[pair]:
            seq_count[keypad_pair] += 1

    # expand through all the keypad conversions:
    for _ in range(num_keypads - 1):
        new_count = dict.fromkeys(ktk_expansions, 0)
        for old_pair, count in seq_count.items():
            for exp_pair in ktk_mappings[old_pair]:
                new_count[exp_pair] += count
        seq_count = new_count
    return sum(seq_count.values())

num_press("029A", 3)
```

```python
# main loop:
num_keypads = 26
sum_compl = 0
for code in codes:
    num_code = int(code[:3])
    compl = num_code * num_press(code, num_keypads)
    sum_compl += compl
print(f'{sum_compl:e}, {sum_compl}')
```
