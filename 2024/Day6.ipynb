In this example, the guard will visit 41 distinct positions on your map.

Predict the path of the guard. How many distinct positions will the guard visit before leaving the mapped area?

```python
with open('input6.txt') as file:
    data = [line.rstrip() for line in file]
grid_width = len(data[0])
grid_height = len(data)
for i, line in enumerate(data):
    try:
        start = (i, line.index('^'))
        break
    except ValueError:
        continue
obstacles = set()
for i, line in enumerate(data):
    for j, char in enumerate(line):
        if char == '#':
            obstacles.add((i, j))

grid_width, grid_height, start, next(iter(obstacles))
```

```python
def outside_grid(pos):
    for coord in pos:
        if coord < 0 or coord >= grid_width: # width == height in input
            return True
    return False
```

```python
from itertools import cycle

directions = cycle(("up", "right", "down", "left"))
facing = next(directions)
visited = set()
visited.add(start)
pos = start
left_grid = False

while True:
    match facing:
        case 'up':
            looking = (pos[0] - 1, pos[1])
        case 'down':
            looking = (pos[0] + 1, pos[1])
        case 'left':
            looking = (pos[0], pos[1] - 1)        
        case 'right':
            looking = (pos[0], pos[1] + 1)
    if outside_grid(looking):
        break
    elif looking in obstacles:
        facing = next(directions)
        continue
    
    pos = looking
    visited.add(pos)
len(visited)
```

It doesn't really matter what you choose to use as an obstacle so long as you and The Historians can put it into position without the guard noticing. The important thing is having enough options that you can find one that minimizes time paradoxes, and in this example, there are 6 different positions you could choose.

You need to get the guard stuck in a loop by adding a single new obstruction. How many different positions could you choose for this obstruction?

```python
# Ok. A few things to note:
# - The new block must be placed on the previous path of the guard (that is be in visited) or it would have no impact whatsoever. so we check all positions in visited minus the starting position
# - when checking, obv we must avoid getting stuck in a loop. How do we recognize we are in a loop? well if we come across the same spot repeatedly. However, visiting the same spot twice in and of itself is not enough, since we can pass it in different directions without getting stuck in a loop.
# - Two ways to address this: We can either 
#       -record the direction we have passed the tile in (prob easiest) or
#       - simply wait until we pass the same spot for the fifth time, then we know for sure (more computeheavy)
# Lets try to implement the first approach
```

```python
def creates_loop(obstacles):
    directions = cycle(("up", "right", "down", "left"))
    facing = next(directions)
    pos = start
    visited = set()    
    visited.add((pos, facing))

    while True:
        match facing:
            case 'up':
                looking = (pos[0] - 1, pos[1])
            case 'down':
                looking = (pos[0] + 1, pos[1])
            case 'left':
                looking = (pos[0], pos[1] - 1)        
            case 'right':
                looking = (pos[0], pos[1] + 1)
        if outside_grid(looking):
            return False
        elif looking in obstacles:
            facing = next(directions)
            continue
        else:
            pos = looking
            if (pos, facing) in visited:
                return True
            else:
                visited.add((pos, facing))
```

```python
obstacle_candidates = visited.copy()
obstacle_candidates.remove(start) # not technically part of the problem definition, but very much undefined
num_loops = 0
for candidate in obstacle_candidates:
    if creates_loop(obstacles.union({candidate})):
        num_loops += 1
num_loops
```
