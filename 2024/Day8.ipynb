Because the topmost A-frequency antenna overlaps with a 0-frequency antinode, there are 14 total unique locations that contain an antinode within the bounds of the map.

Calculate the impact of the signal. How many unique locations within the bounds of the map contain an antinode?

```python
"""
Main challenge: to find antinodes, for any pair of same-type antennas a and b, calculate their distance in 2d space given by b = a + d. Then the two points of the antinodes are a - d and a + 2d.

So we:
- for each possible antenna identifier (upper case letters, lower case letters, digits), collect all antenna locations in a dict
- for each pair of antennas of the same type, calculate the locations of the antinodes
- for each antinode check if it is inbounds of the grid
- count unique antinode locations (two antinodes in the same location count as one point)

"""
```

```python
with open('input8.txt') as file:
    data = [line.rstrip() for line in file]
if len(data) == len(data[0]):
    grid_size = len(data)
grid_size
```

```python
antennas = {}
for i, line in enumerate(data):
    for j, char in enumerate(line):
        if char.isalnum():
            antennas.setdefault(char, []).append((i, j))
antennas
```

```python
def extend_antinode(ant_a, ant_b):
    (a_x, a_y), (b_x, b_y) = ant_a, ant_b
    d_x = b_x - a_x
    d_y = b_y - a_y
    anti_x = b_x + d_x
    anti_y = b_y + d_y
    return(anti_x, anti_y)

def inbounds(coord):
    for c in coord:
        if not 0 <= c <= grid_size - 1:
            return False
    return True
```

```python
extend_antinode((41, 25), (38, 21))
```

```python
antinode_locations = set()
for loc_by_type in antennas.values():
    for ix, ant_a in enumerate(loc_by_type[:-1]):
        for ant_b in loc_by_type[ix+1:]:
            a1 = extend_antinode(ant_a, ant_b)
            a2 = extend_antinode(ant_b, ant_a)
            if inbounds(a1):
                antinode_locations.add(a1)
            if inbounds(a2):
                antinode_locations.add(a2)
len(antinode_locations)
```

```python
# One question I have after reading the assignment is if the antinodes have to be multiples of the distances between antennas 1 and 2, or can be a fraction as well *as long as the fraction leaves them exactly in line* (so the distance 4,4 will have antinodes at every 1,1 point, while the distance 4,3 will not since it is not divisible without decimals)
# The framing of the assignment "exactly in line" sounds like the answer is the former (find lowest divisor in natural numbers) but there is no example of that in the assignment and the implementation of the latter is simpler. so im gonna start with the latter
```

```python
def extend_antinode(ant_a, ant_b):
    (a_x, a_y), (b_x, b_y) = ant_a, ant_b
    d_x = b_x - a_x
    d_y = b_y - a_y
    anti_x = b_x + d_x
    anti_y = b_y + d_y

    if inbounds((anti_x, anti_y)):
        if more_nodes := extend_antinode(ant_b, (anti_x, anti_y)):
            return (more_nodes + [(anti_x, anti_y)])
        else:
            return([(anti_x, anti_y)])
    return []
```

```python
extend_antinode((20,25), (25,18))
```

```python
antinode_locations = set()
for loc_by_type in antennas.values():
    for ix, ant_a in enumerate(loc_by_type[:-1]):
        for ant_b in loc_by_type[ix+1:]:
            antinodes = extend_antinode(ant_a, ant_b) + extend_antinode(ant_b, ant_a)
            if antinodes:
                antinode_locations.update(antinodes)
len(antinode_locations)
```
