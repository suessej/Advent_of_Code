```python
# make one-substitutions easier:
blocks = [[[c for c in row] for row in b] for b in blocks]
def flipped(block):
    return [list(c) for c in zip(*block)]
```

```python
from copy import deepcopy

def sym_id(block):
    syms = set()
    for ix in range(len(block) - 1):
        if all(r1 == r2 for r1, r2 in zip(block[ix::-1], block[ix+1:])):
            syms.add(ix)
    return syms

def fix(o_b):
    # find original id's:
    orig_h_lines = sym_id(o_b)
    orig_v_lines = sym_id(flipped(o_b))
    
    # mutate and match:
    b = deepcopy(o_b)
    for r_ix, r in enumerate(b):
        for c_ix, x in enumerate(r):
            b[r_ix][c_ix] = '#' if x == '.' else '.'
            h_hits = sym_id(b) - orig_h_lines
            v_hits = sym_id(flipped(b)) - orig_v_lines
            if h_hits or v_hits:
                val = 100 * sum(x + 1 for x in h_hits) + sum(x + 1 for x in v_hits)
                return val
            b[r_ix][c_ix] = '.' if x == '.' else '#'
    raise ValueError

# disclosure: Claude helped with bug-fixing, I didn't appreciate the importance of excluding lines from part 1
```
