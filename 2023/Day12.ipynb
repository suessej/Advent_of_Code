```python
# let's start with building a verifier:
import re

def is_consistent(record, grouping):
    damaged = re.findall(r'#+', record)
    if len(damaged) == len(grouping):
        if all(len(r) == g for r, g in zip(damaged, grouping)):
            return True
    return False
is_consistent('#....######.#####.', (1,6,5))
```

```python
# next a permutator:
from itertools import combinations

def num_solutions(dam_rec, grouping):
    tot_hashes = sum(grouping)
    num_q = len([1 for c in dam_rec if c == '?'])
    d_hashes = tot_hashes - len([1 for c in dam_rec if c == '#'])
    
    combos = combinations(range(num_q), d_hashes)
    combos = [''.join('#' if i in co  else '.' for i in range(num_q)) for co in combos]
    num_s = 0

    for co in combos:
        iter_co = iter(co)
        cand = re.sub(r'\?', lambda x: next(iter_co), dam_rec)
        num_s += is_consistent(cand, grouping)
    return num_s
num_solutions('?????##???????.?#??#', (2,2,4,1,1,2))
```

```python
from functools import cache
from math import comb
from collections import deque

def parse_str(rec):
    blocks = re.finditer(r'\?+|#+', rec)
    gr = []
    for b in blocks:
        if b[0].startswith('#'):
            gr.append(len(b[0]))
        elif b[0].startswith('?'):
            left_ext = right_ext = False
            if b.start() > 0 and rec[b.start() - 1] == '#':
                left_ext = True
            if b.end() < len(rec) and rec[b.end()] == '#':
                right_ext = True
            gr.append((left_ext, right_ext, len(b[0])))
    return tuple(gr)
parse_str('.??..??...?##.')
```

```python
# Let's build up our set of possibilities like this:
# If hashes on left and right, one possibility is combination with value L + R + q_len
# If hashes on either side, one possibility is filling up with up to q_len hashes for value of L/R + q_len
# We can also fill it up with smaller values, leaving x leftover-dots
# For the purposes of building new heaps, we are going to assume we have a sequence of question marks boundaried by white dots (or start/end) on both sides (if not we subtract a boundary dot as shown above). Now each additional heap will require one additional boundary dot, so eg if we have . boundaried ??? with len x, we can have one heap len x, two heaps with combined len x-1, three with combined len x-2 etc up to a maximum of len(q) >= 2n - 1 size 1 heaps
# Now leftover dots are given by the reverse: if within a .-boundaried q-space we have n heaps of combined size m, the number of leftover dots is o = len(q) - m - (n-1). The number of places they can attach to is 2 (on the left and on the right) + n-1 (the number of inside boundaries). 
# For the number of permutations where the leftover dots can be placed then we have stars and bars - drawing o dots from n + 1 possible placements with replacement, giving comb(o + n, o)

@cache
def num_comb(q, seq, right_connect=False):
    l_ext, r_ext, q_len = q
    assert not l_ext
    if right_connect: # with right-connect, the last seq_value hugs the right hand side
        q_len -= (seq[-1] + 1)
        seq.pop()
    elif r_ext:
        q_len -= 1

    free_dots = q_len - sum(seq) - len(seq) + 1
    return comb(free_dots + len(seq), free_dots)
```

```python
# Let's look at the heap making/sizing logic in more detail. Roughly go from left to right. We have to do validation: If rec is incompatible with sol, we return 0. We don't have to do the full calc, we can return sub-strings to be popped into the q
# In detail, we pop an element from the record. If it is a #, we check if we need to grow it. If yes, we try to grow it from the next ? segment and hand it back to the q, if we cannot we return 0. If we don't need to (read: must not) grow it, we procceed to the next segment
# By contrast, if our element is a ? segment, we first look at extending a heap to the left. We don't actually need to extend, since that logic was covered previously. But we do need to check if we need to add a . on the left, that is close the heap to the left (remove left_ext, and decrease value by 1).
# Next, we check if it is right-extendable. If yes, we try out all reasonable values to extend to the right. 
# Next, we fill out the ???. On one end, we take the null option (simply delete the segment) and pass that back to the q (). If the segment is right extendable, we instead extend the right segment to the needed length. On the other hand, we try to make a new heap with the value from the solution and pass that back. Let's try to rely on caching here and not worry too much about doing efficient substitutions before exploding the tree. Im not sure if caching will be possible or we have too many combinations but let's see

@cache
def build_paths(rec, sol, pre_count=1):
    if not rec or not sol:
        if sol:
            return None
        if sum([x for x in rec if isinstance(x, int)]):
            return None
        return pre_count
    r_q = deque(rec)
    s_q = deque(sol)

    r = r_q.popleft()
    if isinstance(r, int): # popped in heap
        s = s_q.popleft()
        if (d:=s-r) < 0:
            return None
        elif d == 0:
            return [(tuple(r_q), tuple(s_q), pre_count)]
        else: # see if we can fill it up from the right
            l_heap = r
            if not r_q:
                return None
            r = r_q.popleft()
            if isinstance(r, tuple) and r[0]:
                if r[2] >= d + 2: # we can extend the heap and have leftover-qs
                    r_q.appendleft((False, r[1], r[2] - d - 1))
                    return [(tuple(r_q), tuple(s_q), pre_count)]
                elif r[2] == d + 1: # we can extend the heap without leftover-qs
                    return [(tuple(r_q), tuple(s_q), pre_count)]
                elif r[2] == d: # only soluble if r doesn't wrap to the right
                    if not r[1]:
                        return [(tuple(r_q), tuple(s_q), pre_count)]
                elif r[2] < d: # only potentially soluble if r does wrap to the right (fill and combine)
                    if r[1]:
                        next_heap = r_q.popleft()
                        l_heap += r[2] + next_heap
                        r_q.appendleft(l_heap)
                        s_q.appendleft(s)
                        return [(tuple(r_q), tuple(s_q), pre_count)]
            return None
    else: # start with q_seq
        if r[0]: # close left side since any left side mismatches should've been corrected on previous passes
            if r[2] > 1:
                r_q.appendleft((False, r[1], r[2] - 1))
            return [(tuple(r_q),tuple(s_q), pre_count)]
        if not r[1]: # no right extend
            sum_heaps = 0
            s_left = deque()
            s_right = s_q.copy()
            n_div = -1
            return_values = [(tuple(r_q), tuple(s_q), pre_count)]
            while s_right:
                heap = s_right.popleft()
                s_left.append(heap)
                sum_heaps += heap
                n_div += 1
                space_need = n_div + sum_heaps
                if space_need > r[2]:
                    break
                return_values.append((tuple(r_q), tuple(s_right), num_comb(r, tuple(s_left)) * pre_count))
            return return_values
        else: # right extend possible - fill to the right with various values and close it up 
            return_values = []
            space_need = 0
            max_fill = 0
            for s in s_q:
                possible_fill = min(s - r_q[0], r[2] - space_need)
                max_fill = max(max_fill, possible_fill)
                space_need += 1 + s
                if space_need >= r[2]:
                    break
            for f in range(0, max_fill + 1):
                mod_r_q = r_q.copy()
                mod_r_q[0] = r_q[0] + f
                mod_q = (r[0], False, r[2] - f - 1)
                if mod_q[2] > 0:
                    mod_r_q.appendleft(mod_q)
                return_values.append((tuple(mod_r_q), tuple(s_q), pre_count))
            return return_values 
num_solutions(parse_str('.??..??...?##.'), (1,1,3))
```
