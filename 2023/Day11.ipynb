Expand the universe, then find the length of the shortest path between every pair of galaxies. What is the sum of these lengths?

```python
with open('input11.txt') as file:
    data = [l.rstrip() for l in file]

def flip(table):
    flipped = [[] for _ in table[0]]
    for line in table:
        for ix, c in enumerate(line):
            flipped[ix].append(c)
    for ix, row in enumerate(flipped):
        flipped[ix] = ''.join(row)
    return flipped

def expand_rows(table):
    exp = []
    for line in table:
        exp.append(line)
        if '#' not in line:
            exp.append(line)
    return exp

def expand_all(table):
    return flip(expand_rows(flip(expand_rows(table))))
```

```python
sample_str = '''...#......
.......#..
#.........
..........
......#...
.#........
.........#
..........
.......#..
#...#.....'''
sample = [l for l in sample_str.splitlines()]
exp_sample = expand_all(sample)
```

```python
exp = expand_all(data)
galaxies = []
for i, line in enumerate(exp):
    for j, pos in enumerate(line):
        if pos == '#':
            galaxies.append((i,j))

sum_short = 0
for ix, gal1 in enumerate(galaxies):
    for gal2 in galaxies[ix+1:]:
        dist = abs(gal1[0] - gal2[0]) + abs(gal1[1] - gal2[1])
        sum_short += dist
sum_short
```

--- Part Two ---
The galaxies are much older (and thus much farther apart) than the researcher initially estimated.

Now, instead of the expansion you did before, make each empty row or column one million times larger. That is, each empty row should be replaced with 1000000 empty rows, and each empty column should be replaced with 1000000 empty columns.

(In the example above, if each empty row or column were merely 10 times larger, the sum of the shortest paths between every pair of galaxies would be 1030. If each empty row or column were merely 100 times larger, the sum of the shortest paths between every pair of galaxies would be 8410. However, your universe will need to expand far beyond these values.)

Starting with the same initial image, expand the universe according to these new rules, then find the length of the shortest path between every pair of galaxies. What is the sum of these lengths?

```python
exp_rows = [ix for ix, row in enumerate(data) if '#' not in row]
exp_cols = [ix for ix, col in enumerate(zip(*data)) if '#' not in col]
exp_rows, exp_cols
```

```python
galaxies = []
for i, line in enumerate(data):
    for j, pos in enumerate(line):
        if pos == '#':
            galaxies.append((i,j))
```

```python
sum_short = 0
for ix, gal1 in enumerate(galaxies):
    for gal2 in galaxies[ix+1:]:
        i1, j1 = gal1
        i2, j2 = gal2
        empties = sum([1 for r in exp_rows if min(i1, i2) < r < max(i1,i2)]) + sum([1 for c in exp_cols if min(j1, j2) < c < max(j1,j2)])
        dist = abs(gal1[0] - gal2[0]) + abs(gal1[1] - gal2[1]) + empties * (1000 ** 2 - 1)
        sum_short += dist
sum_short
```
